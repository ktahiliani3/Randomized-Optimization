```python
%matplotlib inline
%matplotlib notebook
%load_ext autoreload
%autoreload 2
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('GA_100_50_50_LOG.csv')
data2 = pd.read_csv('RHC_LOG.csv')
data3 = pd.read_csv('SA_LOG.csv')

iterations = data1.iloc[:,0]

Time1 = data1.iloc[:,5]
Time2 = data2.iloc[:,5]
Time3 = data3.iloc[:,5]

Time1 = np.asarray(Time1)
Time2 = np.asarray(Time2)
Time3 = np.asarray(Time3)
```

```python
plt.plot(iterations, Time1, label = 'Genetic Algorithms')
plt.plot(iterations, Time2,label = 'Randomized Hill Climbing')
plt.plot(iterations, Time3,label = 'Simulated Annealing')
  
# naming the x axis 
plt.xlabel('Iterations') 
# naming the y axis 
plt.ylabel('Time') 
  
# giving a title to my graph 
plt.title('Time Comparison')

plt.legend()
  
# function to show the plot 
plt.show() 
```
