```python
%matplotlib inline
%matplotlib notebook
%load_ext autoreload
%autoreload 2
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('GA_100_50_50_LOG.csv')
iterations = data.iloc[:,0]
MSE_train = data.iloc[:,1]
MSE_test = data.iloc[:,2]
Acc_Train = data.iloc[:,3]
Acc_Test = data.iloc[:,4]
Time = data.iloc[:,5]
iterations = np.asarray(iterations)
MSE_train = np.asarray(MSE_train)
MSE_test = np.asarray(MSE_test)
Acc_Train = np.asarray(Acc_Train)
Acc_Test = np.asarray(Acc_Test)
Time = np.asarray(Time)
```

```python
plt.plot(iterations, MSE_train, label = 'Train')
plt.plot(iterations, MSE_test,label = 'Test')
  
# naming the x axis 
plt.xlabel('Iterations') 
# naming the y axis 
plt.ylabel('MSE') 
  
# giving a title to my graph 
plt.title('Genetic Algorithm - MSE') 

plt.legend()
# function to show the plot 
plt.show() 
```

```python
plt.plot(iterations, Acc_Train, label = 'Train')
plt.plot(iterations, Acc_Test, label = 'Test')
  
# naming the x axis 
plt.xlabel('Iterations') 
# naming the y axis 
plt.ylabel('Accuracy') 
  
# giving a title to my graph 
plt.title('Genetic Algorithm - Accuracy') 

plt.legend()
# function to show the plot 
plt.show() 
```

```python
plt.plot(iterations, Time, label = 'train')
  
# naming the x axis 
plt.xlabel('Iterations') 
# naming the y axis 
plt.ylabel('Time') 
  
# giving a title to my graph 
plt.title('Genetic Algorithm - Time') 
  
# function to show the plot 
plt.show() 
```
