```python
%matplotlib inline
%matplotlib notebook
%load_ext autoreload
%autoreload 2
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('KS_RHC_1_LOG.csv')
data2 = pd.read_csv('KS_SA0.95_1_LOG.csv')
data3 = pd.read_csv('KS_GA100_50_50_1_LOG.csv')
data4 = pd.read_csv('KS_MIMIC300_150_0.9_1_LOG.csv')

iterations1 = data1.iloc[:,0]
iterations2 = data2.iloc[:,0]
iterations3 = data3.iloc[:,0]
iterations4 = data4.iloc[:,0]

fitness1 = data1.iloc[:,1]
fitness2 = data2.iloc[:,1]
fitness3 = data3.iloc[:,1]
fitness4 = data4.iloc[:,1]

time1 = data1.iloc[:,2]
time2 = data2.iloc[:,2]
time3 = data3.iloc[:,2]
time4 = data4.iloc[:,2]


iterations1 = np.asarray(iterations1)
iterations1 = np.asarray(iterations1)
iterations1 = np.asarray(iterations1)
iterations1 = np.asarray(iterations1)

fitness1 = np.asarray(fitness1)
fitness2 = np.asarray(fitness2)
fitness3 = np.asarray(fitness3)
fitness4 = np.asarray(fitness4)

time1 = np.asarray(time1)
time2 = np.asarray(time2)
time3 = np.asarray(time3)
time4 = np.asarray(time4)

```

```python
plt.plot(iterations1, fitness1, label = 'Randomized Hill Climbing',linewidth = 4)
plt.plot(iterations2, fitness2, label = 'Simulated Annealing', linewidth = 4)
plt.plot(iterations3, fitness3, label = 'Genetic Algorithm')
plt.plot(iterations4, fitness4, label = 'MIMIC')
  
# naming the x axis 
plt.xlabel('Iterations') 
# naming the y axis 
plt.ylabel('Fitness Score') 
  
# giving a title to my graph 
plt.title('Knapsack')

plt.legend()
  
# function to show the plot 
plt.show() 
```

```python
plt.plot(iterations1, time1, label = 'Randomized Hill Climbing',linewidth = 3)
plt.plot(iterations2, time2, label = 'Simulated Annealing')
plt.plot(iterations3, time3, label = 'Genetic Algorithm')
plt.plot(iterations4, time4, label = 'MIMIC')
  
# naming the x axis 
plt.xlabel('Iterations') 
# naming the y axis 
plt.ylabel('Time') 
  
# giving a title to my graph 
plt.title('Genetic Algorithm - Time') 

plt.legend()
  
# function to show the plot 
plt.show() 
```
